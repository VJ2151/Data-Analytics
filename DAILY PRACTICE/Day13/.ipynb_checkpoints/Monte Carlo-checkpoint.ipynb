```python
import numpy as np
def  roll_dice():
# simulation of rollng a dice twice,minimum number we fget is 1 or max is 6
# add the results of two simulations,so possible values will be :(1,1)
    return np.sum(np.random.randint(1,7,2))
print(roll_dice())
```

### someone appproches us saying I wil give you 5 dollars if you get 7 and take 1 dollar if you get a number other than 7, how do we know that what will happen ? Our own "Monte carlo Simulation" like function

```python
def monte_carlo_simulation(runs=10000):
    results=np.zeros(2)
    for i in range(runs):
        if roll_dice()==7:
            results[0]+=1
        else:
            results[1]+=1
    return results
```

```python
print(monte_carlo_simulation())
print(monte_carlo_simulation())
print(monte_carlo_simulation())
```
