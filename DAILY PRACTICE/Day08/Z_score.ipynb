```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import zscore
```

```python
df=pd.read_csv("billionaires_2022.csv")
df
```

```python
df['net_worth']=df['net_worth'].str.replace('$','',regex=False)
df['net_worth']=df['net_worth'].str.replace('B','',regex=False)
df['net_worth']=pd.to_numeric(df['net_worth'],errors='coerce')
df=df.dropna(subset=['net_worth'])
```

```python
df
```

```python
import pandas as pd
from scipy.stats import zscore
df['net_worth_zscore'] = zscore(df['net_worth'])
df['net_worth_percentile'] = df['net_worth_zscore'].rank(pct=True) * 100
net_worth_mean = df['net_worth'].mean()
net_worth_std = df['net_worth'].std()
print(f"mean of networth: {net_worth_mean:.2f} B")
print(f"standard deviation of networth: {net_worth_std:.2f} B")
print(df.head())
print(df.tail())
```
