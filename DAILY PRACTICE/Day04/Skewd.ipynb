```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("DAILY PRACTICE/Day04/billionaires.csv")
finalworth=df["finalWorth"]
mean_finalworth=finalworth.mean()
median_finalworth=finalworth.median()
skewness_finalworth=finalworth.skew()
kurtosis_finalworth=finalworth.kurt()
print(mean_finalworth,median_finalworth,skewness_finalworth,kurtosis_finalworth)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(finalworth,kde=True,color="skyblue",bins=30)

plt.axvline(mean_finalworth,color="red",linestyle='--', label = f'mean:{mean_finalworth:.2f}')
plt.axvline(median_finalworth,color="red",linestyle='--', label = f'median:{median_finalworth:.2f}')

plt.title(f'Distribitution of final worth\nskewness:
         {skewnwss_finalworth:.2f},{kurtosis_finalworth:.2f}')
plt.xlabel('final worth')
plt
```
