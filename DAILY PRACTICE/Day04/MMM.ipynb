```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp=[30,31,29,30,32,31,30,29,30,31,30,32,31,30]
prod_sold=[20,22,21,20,23,21,20,22,21,20,22,21,20,100]
df_temp=pd.DataFrame({"Temperature":temp})
df_Sales=pd.DataFrame({"sold product":prod_sold})
print(df_Sales,"\n\n",df_temp)
```

```python
print("mean :",df_temp["Temperature"].mean())
print("median :",df_temp["Temperature"].median())
print("mode :",df_temp["Temperature"].mode()[0])
```

```python
print("mean :",df_Sales["sold product"].mean())
print("median :",df_Sales["sold product"].median())
print("mode :",df_Sales["sold product"].mode()[0])
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))

axes[0].boxplot(df_temp["Temperature"])
axes[0].set_title("Temparature (NO OUTLIER)")
axes[0].set_ylabel("°C")

axes[1].boxplot(df_Sales["sold product"])
axes[1].set_title("Product sold (with outliers)")
axes[1].set_ylabel("Units")
plt.tight_layout()
plt.show()
```
