```python
our_column = df['tip']
our_column = our_column.sample(n=30,replace =False)
meu =3
t_statistics, p_value = stats.ttest_1samp(our_column,meu)
print("PValue: ",p_value)
```

```python
alpha=0.05
if p_value<alpha:
    print("Rejeted H0")
else:
    print(" Do not reject H0")
```

```python
# 2-sample/independent t-test
male_tips = df[df['sex']=='Male']['tip']
female_tips = df[df['sex']=='Female']['tip']

male_tips = male_tips.sample(n=30, replace = False)
female_tips = female_tips.sample(n=30, replace = False)

t_statistics, p_value = stats.ttest_ind(male_tips,female_tips)
print("PValue: ",p_value)

if p_value<alpha:
    print("Rejeted H0")
else:
    print(" Do not reject H0")
```
