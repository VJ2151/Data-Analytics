```python
import numpy as np
import pandas as pd 
```

```python
print("without seed")
print("randint:",np.random.randint(1,10,5))
print("rand:",np.random.rand(5))
print("choice:",np.random.choice([1,2,3,4],1))
```

```python
print("without seed")
print("randint:",np.random.randint(1,10,5))
print("rand:",np.random.rand(5))
print("choice:",np.random.choice([1,2,3,4],5))
```

```python
np.random.seed(8)
print("without seed")
print("randint:",np.random.randint(1,10,5))
print("rand:",np.random.rand(5))
print("choice:",np.random.choice([1,2,3,4],5))
```

```python
np.random.seed(8)
print("without seed")
print("randint:",np.random.randint(1,10,5))
print("rand:",np.random.rand(5))
print("choice:",np.random.choice([1,2,3,4],9))
```

# Uniform and Normal

```python
print("\nuniform:",np.random.uniform(10,50,5))
print("normal:",np.random.normal(50,5,5))
```

```python
companies =np.array(["Infosys","Tcs","HclTech"])
sales =np.array([100,200,300])
profit =np.array([20,50,60])
print("compnaies:",companies)
print("sales:",sales)
print("profit:",profit)
```

```python
print("infosys sales:",sales[0])
print("Infosys profit:",profit[0])
```

```python
sales_s=pd.Series(sales,index=companies)
sales_p=pd.Series(profit,index=companies)
```
