```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b0=-10
b1=0.2
```

```python
weight=np.array([20,30,40,50,60,70])
z=b0+b1*weight
```

```python
def sigmoid(z):
   return 1/(1+np.exp(-z))  
```

```python
probs=sigmoid(z)
probs
print("weight logit(Z) probability result")
for w,logit,p in zip(weight,z,probs):
    result="obese" if p>=0.5 else "not obese"
    print(f"{w:6},{logit:8.2f},{p:10.4f} {result}")
```

```python
x_vals=np.linspace(0,80,200)
z_vals=b0+b1*x_vals
y_vals=sigmoid(z_vals)
```

```python
plt.figure()
plt.plot(x_vals,y_vals,color="blue")
plt.scatter(weight ,probs,color="red")

plt.axhline(y=0.5,color="black")
plt.xlabel("dog weight(kg)",color="green")
plt.ylabel("probability of obesity",color="green")
plt.title("Logistic Regression(dog obesity)",fontdict={'fontweight': 'bold', 'fontsize': 14})
plt.show()
```
