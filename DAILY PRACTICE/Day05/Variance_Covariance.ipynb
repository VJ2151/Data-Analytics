```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("weight-height.csv")
df.head()
```

```python
df_males = df[df['Gender']=='Male']
df_females = df[df['Gender']=='Female']
corr_overall = df[['Height','Weight']]
```

```python
corr_overall = df[['Height','Weight']].corr()
corr_overall
```

```python
corr_male = df_males[['Height','Weight']].corr()
corr_male
```

```python
corr_female = df_females[['Height','Weight']].corr()
corr_female
```

```python
cov_overall = df[['Height','Weight']].cov()
cov_overall
```

```python
cov_male = df[['Height','Weight']].cov()
cov_male
```

```python
cov_female = df[['Height','Weight']].cov()
cov_female
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Overall correlation - Deep Teal
axes[0].scatter(df["Height"], df["Weight"], color='#2A9D8F', alpha=0.5, edgecolor='w', linewidth=0.5)
axes[0].set_title(f'Overall Correlation: {corr_overall.loc["Height", "Weight"]:.2f}')
axes[0].set_xlabel("Height")
axes[0].set_ylabel("Weight")
axes[0].grid(True, linestyle='--', alpha=0.6)

# Male correlation - Steel Blue
axes[1].scatter(df_males["Height"], df_males["Weight"], color='#457B9D', alpha=0.5, edgecolor='w', linewidth=0.5)
axes[1].set_title(f'Male Correlation: {corr_male.loc["Height", "Weight"]:.2f}')
axes[1].set_xlabel("Height")
axes[1].set_ylabel("Weight")
axes[1].grid(True, linestyle='--', alpha=0.6)

# Female correlation - Crimson Rose
axes[2].scatter(df_females["Height"], df_females["Weight"], color='#E63946', alpha=0.5, edgecolor='w', linewidth=0.5)
axes[2].set_title(f'Female Correlation: {corr_female.loc["Height", "Weight"]:.2f}')
axes[2].set_xlabel("Height")
axes[2].set_ylabel("Weight")
axes[2].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Overall correlation - Deep Teal
axes[0].scatter(df["Height"], df["Weight"], color='#2A9D8F', alpha=0.5, edgecolor='w', linewidth=0.5)
axes[0].set_title(f'Overall covariance: {cov_overall.loc["Height", "Weight"]:.2f}')
axes[0].set_xlabel("Height")
axes[0].set_ylabel("Weight")
axes[0].grid(True, linestyle='--', alpha=0.6)

# Male correlation - Steel Blue
axes[1].scatter(df_males["Height"], df_males["Weight"], color='#457B9D', alpha=0.5, edgecolor='w', linewidth=0.5)
axes[1].set_title(f'Male covariance: {cov_male.loc["Height", "Weight"]:.2f}')
axes[1].set_xlabel("Height")
axes[1].set_ylabel("Weight")
axes[1].grid(True, linestyle='--', alpha=0.6)

# Female correlation - Crimson Rose
axes[2].scatter(df_females["Height"], df_females["Weight"], color='#E63946', alpha=0.5, edgecolor='w', linewidth=0.5)
axes[2].set_title(f'Female covariance: {cov_female.loc["Height", "Weight"]:.2f}')
axes[2].set_xlabel("Height")
axes[2].set_ylabel("Weight")
axes[2].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
