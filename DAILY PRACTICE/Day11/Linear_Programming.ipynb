```python
from pulp import *
```

```python
# constraints creation
model=LpProblem("factoryProductionProblem",LpMaximize)
chairs=LpVariable("chairs",0,None,LpInteger)
tables=LpVariable("tables",0,None,LpInteger)
model+=300*chairs+1000*tables,"profit"

model+=5*chairs+10*tables<=400,"wood contraint"
model+=2*chairs+6*tables<=300,"time contraint"

#model+=chairs>=2,"chair contraints"
model.solve()
```

```python
# Each of the variable is printed with its resolved optimum value 
for v in model.variables():
    print(v.name,"=",v.varValue)
print("total profit:",value(model.objective))
```

```python
from pulp import *

# Data
emi_bank_a = 2076
emi_bank_b = 3227
emi_bank_c = 4707
total_loan_required = 10

# Initialize model
model = LpProblem("LoanDistributionProblem", LpMinimize)

# Variables
x1 = LpVariable("Loan_from_bank_A", 0, None, LpContinuous)
x2 = LpVariable("Loan_from_bank_B", 0, None, LpContinuous)
x3 = LpVariable("Loan_from_bank_C", 0, None, LpContinuous)

# 1. ADD THE OBJECTIVE FUNCTION (Missing in your snippet)
model += emi_bank_a * x1 + emi_bank_b * x2 + emi_bank_c * x3, "Total_EMI"

# 2. CONSTRAINTS
model += x1 + x2 + x3 == total_loan_required, "Total_Loan_Constraints"
model += x1 >= 2, "Min_Loan_Bank_A"
model += x3 >= 4, "Min_Loan_Bank_C"  # CHANGED from x1 to x3 to match the name

# Solve
model.solve()

# Output Results
for v in model.variables():
    print(v.name, "=", v.varValue)
print(f"Minimum monthly total emi: {value(model.objective)}")
```
