```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

# line

```python
# average tip across row index-matplotlib
plt.figure(figsize=(6,4))
plt.plot(df.index,df["tip"],marker="o",color="green")
plt.title("line plot-tip across records")
plt.xlabel("Record index")
plt.ylabel("Tip")
plt.show()
```

```python
# line plot using- seaborn 
plt.figure(figsize=(6,4))
sns.lineplot(x=df.index,y=df["tip"],color="green")
plt.title(" Seaborn line plot-tip across records")
plt.xlabel("Record index")
plt.ylabel("Tip")
plt.show()
```

# bar plot


```python
#bar plot  using matplotlib -compare valuus across categories 
avg_bill_day=df.groupby("day")["total_bill"].mean()
#this will create a new dataframe name average_bill,which will contain day and the average of totall bill for that day as columns 
plt.figure(figsize=(6,4))
plt.bar(avg_bill_day.index,avg_bill_day.values,color="skyblue")
plt.title("Matplotlib Bar plot-average total bill by day")
plt.xlabel("day")
plt.ylabel("Average tital bill ")
plt.show()
```

```python
#bar plot using seaborn 
#this will create a new dataframe name average_bill,which will contain day and the average of totall bill for that day as columns 
plt.figure(figsize=(6,4))
sns.barplot(x="day",y="total_bill",color="skyblue",data=df,errorbar=None)
plt.title(" Seaborn plot-average total bill by day")
plt.xlabel("day")
plt.ylabel("Average tital bill")
plt.show()
```

# Histogram

```python
# histogram using matplotlib 
plt.figure(figsize=(6,4))
plt.hist(df["total_bill"],bins=15,color="red")
plt.title("Histogram-Total bill")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# histogram using Seaborn 
plt.figure(figsize=(6,4))
sns.histplot(df["total_bill"],bins=15)
plt.title("Histogram-Total bill")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()
```

# boxplot

```python
# boxplot using matplotlib 
tip_by_day = [df[df["day"] == d] ["tip"] for d in df["day"].unique()]
plt.figure(figsize=(6,4))
plt.boxplot(tip_by_day,tick_labels = df["day"].unique())
plt.title("BOX PLOT-TIP BY DAY")
plt.xlabel("DAY")
plt.ylabel("TIP")
plt.show()
```

```python
# boxplot using seaborn 
plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="tip",data=df)
plt.title("SEABORN PLOT-TIP BY DAY")
plt.xlabel("DAY")
plt.ylabel("TIP")
plt.show()
```

# Scatterd Plot

```python
# scatterd plot-shows realationship between two columns 
plt.figure(figsize=(6,4))
plt.scatter(df["total_bill"],df["tip"],color="skyblue")
plt.title(" Scatterd plot-Total bill vs Tip")
plt.xlabel("total bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill",y="tip",color="skyblue",data=df)
plt.title(" Seaborn Scatterd plot-Total bill vs Tip")
plt.xlabel("total bill")
plt.ylabel("Tip")
plt.show()
```

```python

# Calculate the correlation matrix
corr = df[['total_bill', 'tip', 'size']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap-Correlation Matrix')
plt.show()
```
