# Subplot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(2,3,1)
```

```python
df = sns.load_dataset("tips")
avg_tip_by_day = df.groupby("day")["tip"].mean()
corr = df[["total_bill","tip","size"]].corr()
corr
```

# 1.Traditional/Funcional pyplot style

```python
................3333333plt.figure(figsize=(15,9))

# LINE
plt.subplot(2,3,1)
plt.plot(df.index,df["total_bill"])
plt.title("Line Plot")

# BAR
plt.subplot(2,3,2)
plt.bar(avg_tip_by_day.index,avg_tip_by_day.values,color= "#A6337E")
plt.title("Bar Plot")

# Histogram
plt.subplot(2,3,3)
sns.histplot(df["total_bill"],bins=10,color="#EFC7E2")
plt.title("HISTOGRAM")

# Box
plt.subplot(2,3,4)
plt.boxplot(df["tip"])
plt.title("BOX PLOT")

# SCATTER
plt.subplot(2,3,5)
plt.scatter(df["total_bill"],df["tip"],color="#DB80BD")
plt.title("Scatter plot")

# Heatmap
plt.subplot(2,3,6)
plt.imshow(corr)
plt.title("HeatMap")
plt.colorbar()
plt.show()

plt.tight_layout()
plt.show()
```
