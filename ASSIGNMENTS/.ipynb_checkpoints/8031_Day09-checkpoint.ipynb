```python
import pandas as pd
from datetime import datetime, date
import numpy as np
```

```python
# --------------------------------------------------------------------------------------------------------------
# 1. Write a Pandas program to create
# a) Datetime object for Jan 15 2012.
# b) Specific date and time of 9:20 pm.
# c) Local date and time.
# d) A date without time.
# e) Current date.
# f) Time from a datetime.
# g) Current local time.
# -------------------------------------------------------------------------------------------------------------
a = pd.to_datetime('Jan 15 2012')
print("Datetime object for Jan 15 2012 : ",a)
b = pd.to_datetime('2023-09-20 21:20:00')
print("Specific date and time of 9:20 pm : ",b)
c = datetime.now()
print("Local date and time : ",c)
d = date.today()
print("A date without time : ",d)
e = pd.Timestamp.now().date()
print("Current date : ",e)
f = pd.Timestamp.now().time()
print("Time from a datetime : ",f)
g = pd.Timestamp.now()
print("Current local time : ", g )
```

```python
# 2. Timestamp creations.
a2 = pd.Timestamp('2023-01-01')
b2 = pd.Timestamp('2023-01-01 10:30:00')
c2 = pd.Timestamp.now() + pd.Timedelta(hours=2) # adding time
d2 = pd.Timestamp.now()
print(f"a specific date using timestamp: {a2}\ndate and time using timestamp: {b2}\na time adds in the current local date using timestamp: {c2}\ncurrent date and time using timestamp: {d2} ")
```

```python
# 3. Write a Pandas program to create a date from a given year, month, day and another date from a given string formats.
date_from_comp = pd.Timestamp(year=2022, month=5, day=10)
date_from_str = pd.to_datetime('10-05-2022', format='%d-%m-%Y')
print(date_from_comp,"\n",date_from_str)
```

```python
# 4. Write a Pandas program to print the day after and before a specified date. Also print the days between two given dates.
target = pd.to_datetime('2023-10-10')
yesterday = target - pd.Timedelta(days=1)
tomorrow = target + pd.Timedelta(days=1)
diff = pd.to_datetime('2023-10-20') - target
print(f"target: {target}\nyesterday:{yesterday}\ntomorrow:{tomorrow}\ndiff:{diff}")
```
