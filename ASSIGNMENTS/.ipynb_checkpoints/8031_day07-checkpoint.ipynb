```python
#15. Write a Pandas program to interpolate the missing values using the Linear Interpolation method in a given DataFrame.
#From Wikipedia, in mathematics, linear interpolation is a method of curve fitting using linear polynomials to construct new data points within the 
#range of a discrete set of known data points.
df_interpolated = df.copy()
numeric_cols = df.select_dtypes(include=[np.number]).columns
df_interpolated[numeric_cols] = df[numeric_cols].interpolate(method='linear')

df_interpolated
```

```python
#19. Write a Pandas program to replace the missing values with the most frequent values present in each column of a given DataFrame.
df.fillna(df.mode().iloc[0])
```

```python
#20. Write a Pandas program to create a hitmap for more information about the distribution of missing values in a given DataFrame.
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isna(), cmap='viridis', cbar=True, yticklabels=True,linewidths=0.5,
            linecolor='grey')

plt.title('Heatmap of Missing Values (NaNs) in DataFrame', fontsize=14)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Row Index', fontsize=12)
plt.xticks(rotation=45)
plt.grid(which="minor", color="w", linestyle='-', linewidth=3)
plt.show()
```
