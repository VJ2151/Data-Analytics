## 6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
x = [10, 15, 20, 25, 30]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 2, 2, 2, 2]

plt.plot(x, y1, label='Dashed', linestyle='--')
plt.plot(x, y2, label='Dotted', linestyle=':')
plt.plot(x, y3, label='Dash-dot', linestyle='-.')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Lines with Different Styles')
plt.legend()

plt.show()
```

```python
#9. Write a Python program to display the current axis limits values and set new axis values.
import matplotlib.pyplot as plt

x = [0,10, 20, 30, 40, 50]
y = [0,30, 60, 90, 120, 150]

plt.plot(x, y)

# Displaying current axis limits
print(f"Current axis limits: {plt.axis()}")

# Setting new axis values [xmin, xmax, ymin, ymax]
plt.axis([0, 100, 0, 200])

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Display and Set Axis Limits')

plt.show()
```

```python
#10. Write a Python program to plot quantities which have an x and y position.

x_stars = [2, 3, 5, 6, 8, 9]
y_stars = [1, 5, 10, 18, 20, 25]

x_circles = [3, 4, 7, 10]
y_circles = [5, 12, 15, 22]

plt.plot(x_stars, y_stars, 'b*', label='Stars')
plt.plot(x_circles, y_circles, 'ro', label='Circles')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Star and Circle Quantities')
plt.legend()

plt.show()
```

```python
#12. Write a Python program to create multiple types of charts (a simple curve and plot some quantities) on a single set of axes.

fdata = [
    ("10-03-16", 772.559998),
    ("10-04-16", 776.429993),
    ("10-05-16", 776.469971),
    ("10-06-16", 776.859985),
    ("10-07-16", 775.080017)
]

dates, values = zip(*fdata)

plt.plot(dates, values, color='blue', marker='o')

plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing Stock Value of Alphabet Inc.')

plt.show()
```

```python
#13. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
# Customized the grid lines with linestyle -, width .5. and color blue.

plt.plot(dates, values, color='blue', marker='o')

plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing Stock Value of Alphabet Inc.')

plt.grid(linestyle='-', linewidth='0.5', color='blue')

plt.show()
```

```python
#14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
#Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks.

plt.plot(dates, values, color='red', marker='o')

plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing Stock Value of Alphabet Inc.')

plt.minorticks_on()

plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.tick_params(which='both', top=False, left=False, right=False, bottom=False, 
                labeltop=False, labelleft=True, labelright=False, labelbottom=True)

plt.show()
```
