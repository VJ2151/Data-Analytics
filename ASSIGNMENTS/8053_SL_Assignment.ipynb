```python
import pandas as pd
import seaborn as sns
```

```python
#1. Load the tips dataset into a DataFrame.

df = sns.load_dataset('tips')
df.head()
```

```python
#2. Display the first 5 rows of the tips dataset.

df.head(5)
```

```python
#3. Show all column names in the tips dataset.

df.columns
```

```python
#4. Find the average total_bill.
df['total_bill'].mean()
```

```python
#5. Find the maximum tip given.
df['tip'].max()
```

```python
#6. Filter all rows where total_bill is greater than 20.
df[df['total_bill'] > 20]
```

```python
#7. Filter all rows where smoker is Yes.
df[df['smoker'] == 'Yes']
```

```python
#8. Show only total_bill, tip, and day columns
df[['total_bill','tip','day']]
```

```python
#9. Find the average tip grouped by day.
df.groupby('day')['tip'].mean().reset_index()
```

```python
#11. Count how many records belong to each day.
df['day'].count()
```

```python
#12. Count how many smokers and non-smokers are present.
df['smoker'].value_counts().reset_index()
```

```python
#14. Find the average tip for lunch and dinner separately.
df.groupby('time')['tip'].mean().reset_index()
```

```python
#16. Find the row with the highest tip_percent.
df.loc[df['tip_percent'].idxmax()].reset_index()
```
