## Feature Engineering (66–75)

```python
#66. Create a new column `FamilySize = SibSp + Parch`.
df["Familysize"]=df["sibsp"]+df["parch"]
df
```

```python
#67. Create a new column `IsAlone`.
df["Isalone"]=((df["sibsp"]==0) & (df["parch"]==0)).astype(int)
df
```

```python
#68. Create age groups: child, adult, senior.
df["age"].describe()
```

```python
bins = [0, 18, 60, 100]
labels = ['Child', 'Adult', 'Senior']

# Create the new column
df['AgeGroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
df
```

```python
#70. Extract title from passenger names.
```

```python
#69. Create fare categories: low, medium, high.
bins=[0,15,35,550]
labels=["Low","Medium","High"]
df["FareCategory"]=pd.cut(df["age"], bins=bins, labels=labels)
df
```

```python
df['fare'].describe()
```

```python
#71. Create a column for surname.
```

```python
#72. Create a column `Fare_per_person`.
# 1. Calculate the total family size (including the passenger)
df['FamilySize'] = df['sibsp'] + df['parch'] + 1

# 2. Divide the total fare by the family size
df['Fare_per_person'] = df['fare'] / df['FamilySize']
df
```

```python
#73. Convert `Sex` into numeric values.
gender_map = {'male': 0, 'female': 1}

# Apply the map to the Sex column
df['sex'] = df['sex'].map(gender_map)
df
```

```python
#74. Convert survival values into labels `Yes/No`.
gender_map = {0: "Yes", 1: "No"}

# Apply the map to the Sex column
df['survived'] = df['survived'].map(gender_map)
df
```

```python
#75. Add a column showing age after 5 years.
df["Age_+5"]=df["age"]+5
df
```

## Subplotting Practice (96–100)

```python
#96. Create 2 subplots: age histogram and fare histogram.

plt.subplot(1,2,1)
plt.hist(df["age"], color="#F589E3", edgecolor="black")
plt.title("Histogram of Age")

plt.subplot(1,2,2)
plt.hist(df["fare"], color="#F589E3", edgecolor="black")
plt.title("Histogram of Fare")

plt.tight_layout()
plt.show()
```

```python
#97. Create 2 subplots: survival count and gender count.

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Survival
survived = df['survived'].value_counts()
axes[0].bar(survived.index.astype(str), survived.values, color='orange')
axes[0].set_title('Survival Count')

# Subplot 2: Gender
gender = df['sex'].value_counts()
axes[1].bar(gender.index, gender.values, color='skyblue')
axes[1].set_title('Gender Count')

plt.show()
```

```python
#98. Create 4 subplots: histogram, bar, box, scatter.

fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Top-Left: Histogram
axes[0, 0].hist(df['age'].dropna(), bins=20, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Age Histogram')

# Top-Right: Bar Chart
surv = df['survived'].value_counts()
axes[0, 1].bar(surv.index.astype(str), surv.values, color='salmon')
axes[0, 1].set_title('Survival Bar Chart')

# Bottom-Left: Box Plot
axes[1, 0].boxplot(df['fare'].dropna())
axes[1, 0].set_title('Fare Box Plot')

# Bottom-Right: Scatter Plot
axes[1, 1].scatter(df['age'], df['fare'], alpha=0.5)
axes[1, 1].set_title('Age vs Fare Scatter')

plt.tight_layout()
plt.show()
```

```python
#99. Create subplot using object-oriented syntax.

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

# Use ax[0] for the first plot, ax[1] for the second
ax[0].bar(['No', 'Yes'], df['survived'].value_counts())
ax[0].set_title('Survival')

ax[1].boxplot(df['age'].dropna())
ax[1].set_title('Age Distribution')

plt.tight_layout()
plt.show()
```
