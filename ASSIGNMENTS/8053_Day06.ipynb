## Q1. Display a bar chart of the popularity of programming languages.

**Sample data:**
- Programming languages: Java, Python, PHP, JavaScript, C#, C++
- Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

````python
import matplotlib.pyplot as plt

# Data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create plot
plt.figure(figsize=(8, 5))
# Set zorder=3 to keep bars in front of the grid
plt.bar(languages, popularity, color='Blue', zorder=3)

# Add labels and title
plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Popularity (%)')

# Grid settings
plt.minorticks_on()
# Set zorder=0 to keep the grid in the background
plt.grid(which="minor", linestyle = "--",linewidth = 0.5, zorder=0, color='grey')
plt.grid(which="major", linestyle = "--",linewidth = 0.5,zorder=0, color='red')
# Display
plt.tight_layout()
plt.show()
````

## Q10. Create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

**Sample Data:**
- Means (men)   = (22, 30, 35, 35, 26)
- Means (women) = (25, 32, 30, 35, 29)

````python
men_means   = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)

n = len(men_means)
x = np.arange(n)
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width/2, men_means,   width, label='Men',   color='steelblue')
ax.bar(x + width/2, women_means, width, label='Women', color='salmon')

ax.set_title('Scores by Group and Gender')
ax.set_xlabel('Group')
ax.set_ylabel('Score')
ax.set_xticks(x)
ax.set_xticklabels([f'G{i+1}' for i in range(n)])
ax.legend()
plt.tight_layout()
plt.show()
````

## Q12. Create bar plots with error bars on the same figure.

**Sample Data:**
- Mean velocity:  0.2474, 0.1235, 0.1737, 0.1824
- Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

````python
means = [0.2474, 0.1235, 0.1737, 0.1824]
stds  = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['G1', 'G2', 'G3', 'G4']
x = np.arange(len(labels))

plt.figure(figsize=(8, 5))
plt.bar(x, means, yerr=stds, capsize=5, color='steelblue', alpha=0.7)
plt.xticks(x, labels)
plt.title('Bar Plot with Error Bars')
plt.xlabel('Group')
plt.ylabel('Mean Velocity')
plt.tight_layout()
plt.show()
````

## Q14. Create a stacked bar plot with error bars.

**Note:** Use `bottom` to stack the women's bars on top of the men's bars.

**Sample Data:**
- Means (men)   = (22, 30, 35, 35, 26)
- Means (women) = (25, 32, 30, 35, 29)
- Men Std Dev   = (4, 3, 4, 1, 5)
- Women Std Dev = (3, 5, 2, 3, 3)

````python
men_means    = (22, 30, 35, 35, 26)
women_means  = (25, 32, 30, 35, 29)
men_std      = (4, 3, 4, 1, 5)
women_std    = (3, 5, 2, 3, 3)
groups       = ('G1', 'G2', 'G3', 'G4', 'G5')
x = np.arange(len(groups))

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x, men_means,   label='Men',   color='steelblue', yerr=men_std,   capsize=5)
ax.bar(x, women_means, label='Women', color='salmon',    yerr=women_std, capsize=5,
       bottom=men_means)

ax.set_title('Stacked Bar Plot with Error Bars')
ax.set_xlabel('Group')
ax.set_ylabel('Score')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()
plt.tight_layout()
plt.show()
````

## Q15. Create a horizontal bar chart with differently ordered colors.

**Sample Data:**
```python
languages = [['Language','Science','Math'],
             ['Science','Math','Language'],
             ['Math','Language','Science']]
numbers   = [{'Language':75, 'Science':88, 'Math':96},
             {'Language':71, 'Science':95, 'Math':92},
             {'Language':75, 'Science':90, 'Math':89}]
```

````python
languages = [['Language', 'Science', 'Math'],
             ['Science',  'Math',    'Language'],
             ['Math',     'Language','Science']]

numbers = [{'Language': 75, 'Science': 88, 'Math': 96},
           {'Language': 71, 'Science': 95, 'Math': 92},
           {'Language': 75, 'Science': 90, 'Math': 89}]

color_map = {'Language': 'steelblue', 'Science': 'salmon', 'Math': 'green'}
student_labels = ['Student 1', 'Student 2', 'Student 3']

fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=False)

for i, ax in enumerate(axes):
    subjects = languages[i]
    values   = [numbers[i][s] for s in subjects]
    colors   = [color_map[s] for s in subjects]
    ax.barh(subjects, values, color=colors)
    ax.set_title(student_labels[i])
    ax.set_xlim(60, 100)
    ax.set_xlabel('Score')

plt.suptitle('Horizontal Bar Chart with Differently Ordered Colors')
plt.tight_layout()
plt.show()
````

## Q16. Create a stack bar plot and add label to each section.

**Sample data:**
```python
people   = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4
data = [[ 3.40,  7.71,  6.41, 10.52,  7.53,  7.11, 12.78,  3.45],
        [11.25,  5.04,  6.66, 12.32,  7.46,  6.80,  7.25,  3.69],
        [ 3.94,  4.75, 11.74,  4.65, 13.00,  4.64, 11.17,  8.57],
        [ 4.24, 12.72, 11.38,  9.01, 10.47, 10.98,  3.98,  8.81]]
```

````python
people   = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')
segments = 4
data = [[ 3.40022085,  7.70632498,  6.4097905,  10.51648577,  7.5330039,
           7.1123587,  12.77792868,  3.44773477],
        [11.24811149,  5.03778215,  6.65808464, 12.32220677,  7.45964195,
           6.79685302,  7.24578743,  3.69371847],
        [ 3.94253354,  4.74763549, 11.73529246,  4.6465543,  12.9952182,
           4.63832778, 11.16849999,  8.56883433],
        [ 4.24409799, 12.71746612, 11.3772169,   9.00514257, 10.47084185,
          10.97567589,  3.98287652,  8.80552122]]

colors = ['steelblue', 'salmon', 'green', 'orange']
x = np.arange(len(people))
bottoms = np.zeros(len(people))

fig, ax = plt.subplots(figsize=(10, 6))

for i in range(segments):
    values = data[i]
    bars = ax.bar(x, values, bottom=bottoms, color=colors[i], label=f'Segment {i+1}')
    for bar, val, bot in zip(bars, values, bottoms):
        ax.text(bar.get_x() + bar.get_width() / 2,
                bot + val / 2,
                f'{val:.1f}',
                ha='center', va='center', fontsize=7, color='white', fontweight='bold')
    bottoms += np.array(values)

ax.set_xticks(x)
ax.set_xticklabels(people)
ax.set_title('Stacked Bar Plot with Section Labels')
ax.set_xlabel('Group')
ax.set_ylabel('Value')
ax.legend()
plt.tight_layout()
plt.show()
````

## Q17. Add textures (black and white) to bars and wedges.

**Note:** Use `bottom` to stack the women's bars on top of the men's bars.

````python
men_means   = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
groups      = ('G1', 'G2', 'G3', 'G4', 'G5')
x = np.arange(len(groups))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# --- Stacked Bar with textures ---
bars_men = ax1.bar(x, men_means, label='Men',
                   color='white', edgecolor='black', hatch='//')
bars_women = ax1.bar(x, women_means, bottom=men_means, label='Women',
                     color='white', edgecolor='black', hatch='..')
ax1.set_xticks(x)
ax1.set_xticklabels(groups)
ax1.set_title('Stacked Bar with Textures')
ax1.set_xlabel('Group')
ax1.set_ylabel('Score')
ax1.legend()

# --- Pie chart with hatching ---
pie_data   = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
pie_labels = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
hatches    = ['/', '\\', 'x', 'o', '.', '*']
wedges, texts, autotexts = ax2.pie(
    pie_data, labels=pie_labels, autopct='%1.1f%%',
    colors=['white'] * len(pie_data),
    wedgeprops={'edgecolor': 'black'}
)
for wedge, hatch in zip(wedges, hatches):
    wedge.set_hatch(hatch)
ax2.set_title('Pie Chart with Textures (Wedges)')

plt.suptitle('Black & White Textures on Bars and Wedges', fontsize=13)
plt.tight_layout()
plt.show()
````
