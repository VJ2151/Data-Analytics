```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
df = pd.read_csv("alphabet_stock_data.csv", parse_dates=['Date'], index_col='Date')
df = df.loc['2020-01-01':'2020-12-31']
df.head()
```

##  1. Write a Pandas program to create a line plot of the historical stock prices ofAlphabet Inc. between two specific dates

```python
df['Close'].plot(figsize=(10,5))
plt.title("Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

## 2. Write a Pandas program to create a line plot of the opening, closing stockprices of Alphabet Inc. between two specific dates.

## 3. Write a Pandas program to create a bar plot of the trading volume of AlphabetInc. stock between two specific dates.


```python
plt.figure(figsize=(15, 7))
ax = df['Volume'].plot(kind='bar')
ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=15))

plt.title("Volume Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Volume", fontsize=12)

plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('volume_plot.png')
plt.show()
```

## 14. Write a Pandas program to create a plot of adjusted closing prices, thirtydays and forty days simple moving average of Alphabet Inc. between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

df.index = pd.to_datetime(df.index)
df['SMA_30'] = df['Adj Close'].rolling(window=30).mean()
df['SMA_40'] = df['Adj Close'].rolling(window=40).mean()

df_filtered = df.loc['2020-01-01':'2020-12-31']

df_filtered[['Adj Close', 'SMA_30', 'SMA_40']].plot(figsize=(12, 6))
plt.title("Adjusted Close vs 30 & 40 Day SMA")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

## 15. Write a Pandas program to create a plot of adjusted closing prices, 30 dayssimple moving average and exponential moving average of Alphabet Inc.between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

df.index = pd.to_datetime(df.index)
df['SMA_30'] = df['Adj Close'].rolling(window=30).mean()
df['EMA_30'] = df['Adj Close'].ewm(span=30, adjust=False).mean()

df_filtered = df.loc['2020-01-01':'2020-12-31']

df_filtered[['Adj Close', 'SMA_30', 'EMA_30']].plot(figsize=(12, 6))
plt.title("Adjusted Close vs SMA & EMA")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

## 17. Write a Pandas program to create a plot to visualize daily

```python
import pandas as pd
import matplotlib.pyplot as plt

df.index = pd.to_datetime(df.index)
df['Daily Return'] = df['Close'].pct_change() * 100

df_filtered = df.loc['2020-01-01':'2020-12-31']

df_filtered['Daily Return'].plot(figsize=(12, 6), color='green')
plt.title("Daily Percentage Returns")
plt.ylabel("Percentage Change (%)")
plt.axhline(0, color='black', lw=1)
plt.tight_layout()
plt.show()
```

## 18. Write a Pandas program to plot the volatility over a period of time of Alphabet
Inc. stock price between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df.index = pd.to_datetime(df.index)
df['Daily Return'] = df['Close'].pct_change()
df['Volatility'] = df['Daily Return'].rolling(window=21).std() * np.sqrt(252)

df_filtered = df.loc['2020-01-01':'2020-12-31']

df_filtered['Volatility'].plot(figsize=(12, 6), color='red')
plt.title("Annualized Volatility (21-day Rolling)")
plt.ylabel("Volatility")
plt.tight_layout()
plt.show()
```
