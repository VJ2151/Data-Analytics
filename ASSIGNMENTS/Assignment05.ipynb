## Q49. Create a second DataFrame and concatenate it with the first one.

```python
df2 = pd.DataFrame({
    'Name':       ['Frank', 'Grace'],
    'Age':        [20, 21],
    'Score':      [72, 88],
    'Passed':     [True, True],
    'Department': ['CS', 'IT']
})
df_combined = pd.concat([df, df2], ignore_index=True)
print(df_combined)
```

## Q50. Create two DataFrames with a common column and merge them.

```python
df_a = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'Score': [85, 92, 78]})
df_b = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'City': ['Delhi', 'Mumbai', 'Chennai']})
merged = pd.merge(df_a, df_b, on='Name')
print(merged)
```
