```python
# 14. Write a Pandas program to check if a day is a business day (weekday) or not
def is_business_day(date_str):
    dt = pd.to_datetime(date_str)
# bday.is_on_offset checks if the date falls on a business day
    return pd.offsets.BusinessDay().is_on_offset(dt)
print(f"2024-05-01 (Wed): {is_business_day('2024-05-01')}")
print(f"2024-05-04 (Sat): {is_business_day('2024-05-04')}")
```

```python
import pandas as pd

def check_business_day(date_str):
    dt = pd.to_datetime(date_str)
    # dayofweek < 5 (0 is Monday, 4 is Friday)
    return "Business Day" if dt.dayofweek < 5 else "Weekend"
print(f"2024-05-10: {check_business_day('2024-05-10')}") # Friday
print(f"2024-05-11: {check_business_day('2024-05-11')}") # Saturday
```

```python
# 17. Write a Pandas program to convert unix/epoch time to a regular time stamp in UTC. Also convert the 
# said timestamp in to a given time zone.
import pandas as pd

epoch_time = 1715341200 
# Corrected: Specify unit and UTC true immediately
utc_ts = pd.to_datetime(epoch_time, unit='s', utc=True)
print(f"UTC: {utc_ts}")

# Convert to specific timezone (e.g., 'US/Eastern' or 'Asia/Kolkata')
local_ts = utc_ts.tz_convert('Asia/Kolkata')
print(f"Local: {local_ts}")
```

```python
## 19. Write a Pandas program to remove the time zone information from a Time series data
import pandas as pd

# 1. Create a time series with a time zone (UTC)
ts_aware = pd.date_range('2024-05-01', periods=3, freq='D', tz='UTC')
print("With Time Zone:\n", ts_aware)

# 2. Remove the time zone information
ts_naive = ts_aware.tz_localize(None)
print("\nWithout Time Zone (Naive):\n", ts_naive)
```
