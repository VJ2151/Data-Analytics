#### 28. Write a Pandas program to get the day of month, day of year, week number and day of week from a given series of date strings.

```python

s = pd.Series(['01 Jan 2015', '10-02-2016', '20180307', '2014/05/06', '2016-04-12', '2019-04-06T11:20'])
print("Original Series:")
print(s)
ts = pd.to_datetime(s, format='mixed')
print("\nDay of month:")
print(ts.dt.day.tolist())
print("\nDay of year:")
print(ts.dt.dayofyear.tolist())
print("\nWeek number:")
print(ts.dt.isocalendar().week.astype(int).tolist())
print("\nDay of week:")
print(ts.dt.day_name().tolist())
```

#### 31. Write a Pandas program to compute the Euclidean distance between two given series.Euclidean distanceFrom Wikipedia,In mathematics, the Euclidean distance or Euclidean metric is the "ordinary" straight-linedistance between two points in Euclidean space. With this distance, Euclidean spacebecomes a metric space. The associated norm is called the Euclidean norm.

```python
s = pd.Series([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])

# Use shift to compare current values with neighbors
# shift(1) moves values down (comparing to previous)
# shift(-1) moves values up (comparing to next)
temp = np.where((s > s.shift(1)) & (s > s.shift(-1)))[0]

print("Original series:")
print(s)
print("\nPositions of the values surrounded by smaller values on both sides:")
print(temp)
```

#### 32. Write a Pandas program to find the positions of the values neighboured by smaller values on both sides in a given series.

#### 33. Write a Pandas program to replace missing white spaces in a given string with the least frequent character.

```python
import pandas as pd
text = 'abc def abcdef icd'

char_series = pd.Series(list(text.replace(" ","")))

least_freq_char = char_series.value_counts(ascending=True).idxmin()

result = text.replace(" ", least_freq_char)

print("Original series:")
print(text)
print(char_series.value_counts(ascending=False))
print(result)
```

#### 34. Write a Pandas program to compute the autocorrelations of a given numeric series.From Wikipedia: Autocorrelation, also known as serial correlation, is the correlation of a signal with a delayed copy of itself as a function of delay. Informally, it is the similarity between observations as a function of the time lag between them.

```python
s = pd.Series(np.random.normal(1, 10, 15))
autocorrelations = [round(s.autocorr(lag=i), 2) for i in range(1, 11)]
print("Original series:")
print(s)
print("\nAutocorrelations of the said series:")
print(autocorrelations)
```

#### 40. Write a Pandas program to check inequality over the index axis of a given dataframe and a given series.

```python
# Create Sample DataFrame
df = pd.DataFrame({
    'W': [68.0, 75.0, 86.0, 80.0, np.nan],
    'X': [78.0, 75.0, np.nan, 80.0, 86.0],
    'Y': [84, 94, 89, 86, 86],
    'Z': [86, 97, 96, 72, 83]
})

# Create Sample Series
s = pd.Series([68.0, 75.0, 86.0, 80.0, np.nan])

print("Original DataFrame:")
print(df)
print("\nOriginal Series:")
print(s)

# Check for inequality along the index (axis=0)
result = df.ne(s, axis=0)

print("\nCheck for inequality of the said series & dataframe:")
print(result)
```
