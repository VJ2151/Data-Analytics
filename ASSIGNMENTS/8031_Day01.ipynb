# Write a NumPy program to compute the x and y coordinates for points on a sine curve and plot thepoints using matplotlib.

# Write a NumPy program to replace all numbers in a given array which is equal, less and greater to agiven number.

```python
import numpy as np
data = np.array([10, 20, 30, 40, 50])
target = 30
data[data == target] = 0
data[data < target] = -1
data[data > target] = 1
print(data)
```
